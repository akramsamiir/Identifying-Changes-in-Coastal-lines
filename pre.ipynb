Remove One Elements images

```python
def data_handel(labels_path, images_path):
    valid_labels = []
    
    for image in os.listdir(labels_path):
        image_readed = cv2.imread(labels_path+"/"+image)
        if np.array_equal(image_readed, red_image) or np.array_equal(image_readed, black_image):
            os.remove(labels_path + "/" + image)
            continue
        dot_index = image.index(".")
        valid_labels.append( image[:dot_index] )
        
        
    for image in os.listdir(images_path):
        dot_index = image.index(".")
        if  image[:dot_index] not in  valid_labels:
            os.remove(images_path + "/" + image)
```

```python
data_handel(labels_data_path_rgb, train_data_path_rgb)
```

```python
def flip_image(train_path, labels_path):
    for image in os.listdir(train_path):
        image_obj = Image.open(train_path + "/" +image)
        hori_flippedImage = image_obj.transpose(Image.FLIP_LEFT_RIGHT)
        hori_flippedImage.save(f"{train_path}/{image [  : image.index('.') ]}_flipped.png","png")
        
    for image in os.listdir(labels_path):
        image_obj = Image.open(labels_path + "/" +image)
        hori_flippedImage = image_obj.transpose(Image.FLIP_LEFT_RIGHT)
        hori_flippedImage.save(f"{labels_path}/{image [  : image.index('.') ]}_flipped.png","png")
```

```python

def one_channel(labels_path, folder):
    os.chdir(folder)
    for image in os.listdir(labels_path):
        image_obj = cv2.imread(labels_path + "/" +image)
        edit_image = image_obj[:,:,[2]]
        (thresh, BnW_image) = cv2.threshold(edit_image, 125, 255, cv2.THRESH_BINARY)
        cv2.imwrite(f"{folder}/{image}.png",BnW_image)
```
